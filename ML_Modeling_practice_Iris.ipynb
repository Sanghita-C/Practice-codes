```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
feature_data = pd.DataFrame(iris.data, columns = iris.feature_names)
target_label = pd.DataFrame(iris.target, columns = ['label'])
data = pd.concat([feature_data, target_label], axis = 1)
data.head()
```

```python
print(data.describe())
```

```python
import matplotlib.pyplot as plt
```

```python
data['label'].value_counts().plot(kind = 'bar')
plt.xlabel('labels')
plt.ylabel('counts')
plt.title('label distribution')
plt.show()
```

```python
#checking the relation of features with each other
import seaborn as sns

# Pairplot with hue as the class label
sns.pairplot(data, hue = 'label', diag_kind='hist')
plt.suptitle("Pairwise Feature Dependency", y=1.02)
plt.show()
```

```python
corr_matrix = data.corr()
print(corr_matrix)
```
